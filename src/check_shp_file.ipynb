```python
import pandas as pd
import geopandas as gpd
from pathlib import Path
```

```python
# Defining directories
dir_path = Path.cwd().parent
data_folder = dir_path.joinpath("data/shapes")

lan_path = data_folder.joinpath('lantana/lantana_camara.shp')
shp_path = data_folder.joinpath('himachal_species/himachal_species.shp')

final_folder = data_folder.joinpath("himachal_species_final")
if not final_folder.exists():
    final_folder.mkdir(parents=True)

species_shp_path = final_folder.joinpath('himachal_species_final.shp')

```

```python
shp_df = gpd.read_file(shp_path)
shp_df
```

```python
shp_df['classname'].value_counts()
```

### Removing Himalayan Birch

```python
shp_df = shp_df[shp_df['classname']!="himalayan birch"]
print(shp_df.shape)
shp_df['classname'].value_counts()
```

```python
lan_df = gpd.read_file(lan_path).rename(columns={
    'Latitude':'Lat',
    'Longitude':'Long',
    'Species_Na':'classname',
    'Vegetation':'s_type'

}).drop('Common_Nam', axis=1)
lan_df
```

There are 4 duplicates in lantana file which have been corrected for

```python
lan_df = lan_df.drop_duplicates(['Long', 'Lat'], keep='first')
lan_df
```

```python
lan_df[lan_df.duplicated(['Long', 'Lat'], keep=False)]
```

```python
shp_df.duplicated(['Long', 'Lat'],keep=False).value_counts()
```

```python
species_shp = pd.concat([shp_df, lan_df], axis=0)
species_shp
```

```python
species_shp['classname'].value_counts()
```

```python
species_shp.duplicated(['Long', 'Lat'],keep=False).value_counts()
```

### Renamimg the species


```python
species_shp['classname'].unique()
```

```python
species_dict = {
    "aesculusi indica":"Aesculus indica",
    "senegalia catechu":"Senegalia catechu",
    "cedrus deodara":"Cedrus deodara",
    "indian gooseberry":"Phyllanthus emblica",
    "taxus baccata":"Taxus baccata",
    "rhododendron":"Rhododendron arboreum",
    "chir pine":"Pinus roxburghii",
    "dendrocalamus strictus":"Dendrocalamus strictus",
    "wild mango":"Mangifera sylvatica",
    "shorea robusta":"Shorea robusta",
    "bambusa bamboo":"Bambusoideae spp.",
    "quercus leucotrichophora":"Quercus leucotrichophora",
    "eucalyptus":"Eucalyptus globulus",
    "Lantana camara":"Lantana camara",
}
```

```python
species_shp['classname'] = species_shp['classname'].replace(species_dict)
```

```python
species_shp['classname'].value_counts()
```

```python
species_shp.to_file(species_shp_path, driver='ESRI Shapefile')
```
